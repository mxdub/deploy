```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Stynhtetic data
x1 = np.random.normal(size = 100)
x2 = np.random.normal(2, size = 100)
noise = np.random.normal(0, 1, size = 100)

y = 0.5 +  2*x1 + x2**2 + noise

df = pd.DataFrame({'y':y, 'x1':x1, 'x2':x2})

sns.pairplot(df)
```

```python
y = np.random.binomial(n = 1, p = (-np.min(y) + y + 0.01) / ( (np.max(y) - np.min(y)) + 0.01))
```

```python
from sklearn.linear_model import LogisticRegression

X = pd.DataFrame({'x1':x1, 'x1_2':x1**2, 'x2':x2, 'x2_2':x2**2})

pd.concat([pd.Series(y, name = 'y'), X], axis = 1).to_csv('flask/dataset.csv', index=False)

lm = LogisticRegression().fit(X, y)
lm.score(X,y)

# [lm.intercept_, lm.coef_]
```

```python
import pickle

with open('flask/lm_model.pkl', 'wb') as outp:
    pickle.dump(lm, outp, pickle.HIGHEST_PROTOCOL)
    
```
